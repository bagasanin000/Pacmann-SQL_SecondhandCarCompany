# CREATE USER DUMMY DATA AND MODIFY CITY DUMMY DATA

```python
# preparing library

import pandas as pd
import random
from faker import Faker
fake = Faker('id_ID')
```

```python
# rename nama_kota to city_name

df_city = pd.read_excel("dummy/city.xlsx")
df_city.rename(columns={"nama_kota":"city_name"}, inplace=True)
df_city.head()
```

```python
# take location_id from dummy city data

df_city.rename(columns={"kota_id":"location_id"}, inplace=True)
loc_id_options = df_city["location_id"]
loc_id_options
```

```python
# save city dummy as csv

df_city.to_csv("dummy/cities_table.csv", index=False)
```

```python
# make random list of name, phone, and address

n = 60
name_list = []
phone_list = []
address_list = []

for i in range(0, n):
    name = fake.name()
    name_list.append(name)
    
    phone = fake.phone_number()
    phone_list.append(phone)
    
    address = fake.street_address()
    address_list.append(address)
```

```python
loc_list = []
for i in range(0, n):
    x = random.randint(0, len(loc_id_options)-1)
    loc_list.append(loc_id_options[x])
```

```python
df_user = pd.DataFrame()
df_user['user_id'] = range(1, n+1)
df_user['name'] = name_list
df_user['phone'] = phone_list
df_user['address'] = address_list
df_user['location_id'] = loc_list
```

```python
df_user.head()
```

```python
# save as csv

df_user.to_csv("dummy/users_table.csv", index=False)
```

# MODIFY EXISTING CAR DUMMY DATA

```python
# read car data

df_car = pd.read_excel("dummy/car_product.xlsx")
df_car.head()
```

```python
# create random transmission data

transmission_options = ["manual", "auto", "hybrid"]

transmission_list = []
for i in range(0, 50):
    x = random.randint(0, 2)
    transmission_list.append(transmission_options[x])

df_car['transmission'] = transmission_list
df_car.head()
```

```python
# rename product_id to car_id

df_car.rename(columns={'product_id':'car_id'}, inplace=True)
df_car.head()
```

```python
# create random description

colour_options = ["", "red", "yellow", "blue", "dark blue", "grey", "black", "good condition"]

desc_list = []
for i in range(0, 50):
    x = random.randint(0, len(colour_options)-1)
    desc_list.append(colour_options[x])

df_car['description'] = desc_list
df_car.head()
```

```python
# export as csv

df_car.to_csv("dummy/cars_table.csv", index = False)
```
